```python
import numpy as np
import pandas as pd
from numpy.random import randn

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#inline display
%matplotlib inline
```

```python
# Regression plots using sns. (lmplot)

tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# Lets us lmplot
sns.lmplot("total_bill","tip", tips)
```

```python
# Just like before, we can change the color and other prop of this plot
sns.lmplot("total_bill","tip", tips,
           scatter_kws={"marker":"*", "color":"indianred"},
           line_kws={"linewidth":2,"color":"blue"}
          
          )
```

```python
# Remove a regression from lmplot
sns.lmplot("total_bill","tip",tips,fit_reg=False)
```

```python
# lmplots also works with discrete variables
tips["tip_pect"]=100*(tips['tip'])/tips['total_bill']
```

```python
tips.head()
```

```python
sns.lmplot("size","tip_pect",tips)
```

```python
# Jitter http://en.wikipedia.org/wiki/Jitter
sns.lmplot("size","tip_pect", tips, x_jitter=.1)
```

```python
# Hue, to define subsets
sns.lmplot("total_bill","tip_pect",tips, hue="sex", markers=["x","o"])
```

```python
sns.lmplot("total_bill","tip_pect",tips, hue="day")
```

```python
# Finally, regplot
sns.regplot("total_bill","tip_pect", tips)
```
